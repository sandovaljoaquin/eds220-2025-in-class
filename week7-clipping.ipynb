## Clipping 

- using a polygon(s) to only select data within in. 

Syntax: 
```
updated_geof = geopandas.clip(geodf, mask)
```

where: 
updated_geodf = intsection of the geometries in geodf with mask 
- mask = gpd.GeoDataFrame with polygons you want for clipping 

Male sure CRS for mask and geodf match 

````python
# Clip populated places to Alaska multipolygon

ak_places = gpd.clip(places, alaska)

fig, ax = plt.subplots()

alaska.plot(ax = ax)
ak_places.plot(ax = ax, color = 'red')

plt.show()
````

````python
print('CRS: ', roads.crs)
roads.head()
````

Before jumping into reprojecting and clipping, we will subset the data to only US roads. 

Good practice to reduce geospatial data as much as possible before applying geospatial transformations 

````python
usa_roads = roads[roads.sov_a3 == 'USA']
usa_roads.plot()
````

````python
#roads = roads.to_crs('epsg:3338')

#assert roads.crs == '3338'
````

````python

# Clip usa_roads to alaska geometry
ak_roads = gpd.clip(usa_roads.to_crs(alaska.crs), alaska)


fig, ax = plt.subplots()
alaska.plot(ax=ax)
ak_roads.plot(ax=ax, color='red')
plt.show()
````

## Clipping with a bounding box 

We will clip the usa_roads geo-df with the bounding box of alaska instead of the polygons 

We will create the bounding box with the box() function from shapely

we construct a box by passing 4 numbers into 'box()'

`box(minx, miny, maxx, maxy)'

````python
alaska.total_bounds 
````

````python
box(-2175729.73162241,   404545.10791528,  1493082.30904445, 2383808.43520815)
````

````python
bbox = box(*alaska.total_bounds)        
print(type(box))
bbox
````

`*` = unpacking operator

Think about it as "unpacking" the elements in `alaska.total_bounds` and assigning them one by one to the parameters `minx`, `miny`, `maxx`, `maxy` of the `box()` function. 

````python
# Create a geo-dataframe from Alaska bounding box 

ak_bbox = gpd.GeoDataFrame(geometry = [bbox],    # Assign geometry column
                          crs = alaska.crs)     # Assign CRS
````

````python
ak_bbox.plot()
````

Clip the roads using Alaska's bounding box

````python
ak_complete_roads = gpd.clip(usa_roads.to_crs(ak_bbox.crs), ak_bbox)
````

````python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,10))

ak_roads.plot(ax=ax1)
ax1.set_title('Roads clipped with AK multipolygon')

ak_complete_roads.plot(ax=ax2)
ax2.set_title('Roads clipped with AK bounding box')

plt.show()
````

````python
fig, ax = plt.subplots(figsize=(11,5))
ax.axis('off')

alaska.plot(ax = ax, color = 'whitesmoke', edgecolor = '0.7')

ak_complete_roads.plot( ax = ax, 
                       zorder = 1,      # Specify layer plotting order 
                       column = 'type', 
                       legend = True, 
                       legend_kwds={'title': "Road Types", 
                                    'loc': 'upper left',
                                    'bbox_to_anchor':(0,0.9),
                                    'fontsize':'small'})

ak_places.plot( ax = ax,
               zorder = 1, 
               color = 'red', 
               marker = 's' )
        
# Add city names as text annotations

for x, y, name in zip(ak_places.geometry.x, ak_places.geometry.y, ak_places['name']):
    if name == 'Valdez': 
            ax.text(x+60000, y+55000, name, fontsize=8, ha='right') 
    else:   ax.text(x-30000, y+20000, name, fontsize=8, ha='right') 
ax.set_title("Road Networks and Major Cities in Alaska", fontsize=14, fontweight='bold')
plt.show()
````
